```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
inner_loss_list = []
with open("inner_loss.txt", "r") as file:
    for line in file:
        line = line.rstrip("\n")
        inner_loss_list.append(float(line))
```

```python
sum(inner_loss_list) / len(inner_loss_list)
```

```python
# 创建直方图
plt.hist(inner_loss_list, edgecolor='white')

# 添加标题和标签
plt.title('Inner Loop MSE Loss (per batch) Frequency Histogram')
plt.xlabel('Inner Loop MSE Loss (per batch)')
plt.ylabel('Frequencies')

# 显示图形
plt.show()
```

```python
psnr_list = []
with open("psnr.txt", "r") as file:
    for line in file:
        line = line.rstrip("\n")
        psnr_list.append(float(line))
```

```python
sum(psnr_list) / len(psnr_list)
```

```python
# 创建直方图
plt.hist(psnr_list, edgecolor='white', color='orange')

# 添加标题和标签
plt.title('Peak Signal-to-Noise Ratio (PSNR per batch) Frequency Histogram')
plt.xlabel('PSNR (per batch)')
plt.ylabel('Frequencies')

# 显示图形
plt.show()
```
